```python
from openai import AzureOpenAI
import dotenv
import os
import base64
from azure.core.credentials import AzureKeyCredential

dotenv.load_dotenv(override=True)
```

```python
client = AzureOpenAI(
    azure_endpoint=os.getenv('AzureOpenAiEndpoint'),
    api_key=os.getenv('AzureOpenAiKey'),
    azure_deployment='gpt-4o',
    api_version='2024-02-15-preview'
)
```

```python
with open("table.png", "rb") as image_file:
    image_data = base64.b64encode(image_file.read()).decode('utf-8')
```

```python
image_data
```

```python
def analyze_document(encoded_image,prompt):
    res = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": prompt },
                    {
                        "type": "image_url",
                        "image_url": {"url": f"data:image/png;base64,{encoded_image}"}
                    },
                ],
            }
        ],
    )
    return res.choices[0].message.content
```

```python
prompt = 'What is the Accuracy in the Results Column for Dexterity?'
analyze_document(image_data, prompt)
```

```python
prompt = "What is the Accuracy in the Results Column for Low Vision?"
analyze_document(image_data, prompt)
```

```python
prompt = "What is the Time to complete in the results column for low vision?"
analyze_document(image_data, prompt)
```

```python
prompt = "How many participants are covered?"
analyze_document(image_data, prompt)

```
